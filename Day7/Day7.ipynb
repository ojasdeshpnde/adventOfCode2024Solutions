```python
import math
data = []
with open("inputd7.txt", "r") as file:
    for line in file:
        data.append(line.strip())
```

```python
def prettifyData(data):
    retValue = []
    for row in data:
        localList = []
        line = row.split(":")
        result = int(line[0])
        nums = [int(i) for i in line[1].strip().split(" ")]
        localList.append(result)
        localList.append(nums)
        retValue.append(localList)
    return retValue

# 0s are '+' signs and 1s are '*' signs

def validateExpression(operators, data, result):
    total = data[0]
    for i in range(len(operators)):
        if(operators[i] == 0):
            total += data[i+1]
        elif(operators[i] == 1):
            total *= data[i+1]
    return total == result

def generateOperators(data, index):
    numOfOperators = len(data) -1
    retValue = []
    for i in range(numOfOperators):
        if(index - pow(2,numOfOperators - i - 1)>=0):
            index -= pow(2,numOfOperators - i - 1)
            retValue.append(1)
        else:
            retValue.append(0)
    return retValue

def iterator(rawData):
    data = prettifyData(rawData)
    retValue = 0
    for row in range(len(data)):
        result = data[row][0]
        nums = data[row][1]

        for i in range(pow(2,len(nums)-1)):
            currOps = generateOperators(nums,i)
            if(validateExpression(currOps,nums,result)):
                retValue += result
                break
    return retValue
```

```python
print(iterator(data))
```

```python
def validateExpressionAdvanced(operators, data, result):
    total = data[0]
    for i in range(len(operators)):
        if(operators[i] == 0):
            total += data[i+1]
        elif(operators[i] == 1):
            total *= data[i+1]
        elif(operators[i] == 2):
            total = int(str(total) + str(data[i+1]))
    return total == result

def generateOperatorsAdvanced(data, index):
    numOfOperators = len(data) -1
    retValue = []
    for i in range(numOfOperators):
        if(index - 2*pow(3,numOfOperators - i - 1)>=0):
            index -= 2*pow(3,numOfOperators - i - 1)
            retValue.append(2)
        elif(index - pow(3,numOfOperators - i - 1) >= 0):
            index -= pow(3,numOfOperators -1 -i)
            retValue.append(1)
        else:
            retValue.append(0)
    return retValue

def iteratorAdvanced(rawData):
    data = prettifyData(rawData)
    retValue = 0
    for row in range(len(data)):
        result = data[row][0]
        nums = data[row][1]
        for i in range(pow(3,len(nums)-1)):
            currOps = generateOperatorsAdvanced(nums,i)
            if(validateExpressionAdvanced(currOps,nums,result)):
                retValue += result
                break
    return retValue
```

```python
print(iteratorAdvanced(data))
```
